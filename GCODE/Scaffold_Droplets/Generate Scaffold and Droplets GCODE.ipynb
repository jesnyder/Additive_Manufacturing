```python
# Generate the GCODE to print a square scaffold.

import time

length = 20; # Printed construct length
center_to_center = 1; # Center to center spacing of printed struts
rows = int(length/center_to_center); # Number of rows
Etime = 1; 

Zhighpoint = -4;
Zlowpoint = 0;
Edrop = 0.1;

timestr = time.strftime("%Y%m%d-%H%M%S")

text_file = open('Code Generated/ScaffoldandDroplets_' + timestr + '_NOT_VALIDATED_.gcode', "a+")
text_file.write("\n; Date Compiled: %s" % timestr )
text_file.write("\n; GCODE Reference: http://reprap.org/wiki/G-code#G0_.26_G1:_Move")
text_file.write("\nG21; set units to millimeters")
text_file.write("\nG90; use G90 for absolute and G91 for incremental/relative coordinates")
text_file.write("\nG83; use G82 for absolute and G83 for incremental/relative distance for extrusion")

text_file.write("\n \nT0 ; Nozzle 0")
text_file.write("\n \nG1 E28 ; Nozzle 0")

j=1;
while j<=2: 
    
    E1 = Etime*center_to_center;
    E2 = Etime*length;

    i = 0; 
    while i<(int(rows)): 
     
        if j ==1: 
            x1 = i*center_to_center; x2 = x1;
            y1 = 0; y2 = length;
        else:
            x1 = 0; x2 = length;
            y1 = i*center_to_center; y2 = y1;
    
        text_file.write("\n")
        text_file.write("\nG1 X%.2f" % x1)
        text_file.write(" Y%.2f" % y1)     
        text_file.write(" Z%.2f" % 0)
        text_file.write(" E%.2f" % E1)
    
        text_file.write("\nG1 X%.2f" % x2)
        text_file.write(" Y%.2f" % y2)     
        text_file.write(" Z%.2f" % 0)
        text_file.write(" E%.2f" % E2)
    
        i = i+1;
        
        if j ==1: 
            x1 = i*center_to_center; x2 = x1;
            y1 = length; y2 = 0;
        else:
            x1 = length; x2 = 0;
            y1 = i*center_to_center; y2 = y1;

        text_file.write("\nG1 X%.2f" % x1)
        text_file.write(" Y%.2f" % y1)     
        text_file.write(" Z%.2f" % 0)
        text_file.write(" E%.2f" % E1)
        
        text_file.write("\nG1 X%.2f" % x2)
        text_file.write(" Y%.2f" % y2)     
        text_file.write(" Z%.2f" % 0)
        text_file.write(" E%.2f" % E1)
    
        i =i+1;
    
    text_file.write("\n \nG1 X%.2f" % 0)
    text_file.write(" Y%.2f" % 0)     
    text_file.write(" Z%.2f" % -3)
    text_file.write(" E%.2f" % E2)
    
    j=j+1;
    

text_file.write("\n \nT1 ; Nozzle 0")
text_file.write("\n \nG1 E20 ; Nozzle 0")
     
j = 0; 
while j <(int(rows)):
    
    y = j*center_to_center+0.5*center_to_center
    
    i = 0; 
    while i<(int(rows)):
        
        x = i*center_to_center+0.5*center_to_center
    
        text_file.write("\n")
        text_file.write("\n G1 X%.2f" % x)
        text_file.write(" Y%.2f" % y)     
        text_file.write(" Z%.2f" % Zhighpoint)
        text_file.write(" E0")
    
        text_file.write("\n G1 X%.2f" % x)
        text_file.write(" Y%.2f" % y)   
        text_file.write(" Z%.2f" % Zlowpoint)
        text_file.write(" E%.2f" % Edrop) 
    
        text_file.write("\n G1 X%.2f" % x)
        text_file.write(" Y%.2f" % y)     
        text_file.write(" Z%.2f" % Zhighpoint)
        text_file.write(" E0")
        
        i=i+1;
        
    j=j+1;

#text_file.write("\n G1 Z%.2f" % 5*Zhighpoint)
text_file.close()

```
